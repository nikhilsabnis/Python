```python
from random import shuffle
```

```python
def shuffle_list(list_Values):
    shuffle(list_Values)
    return list_Values
```

```python
def get_user_guess():
    guess_input = input("Choose position of the ball is (0,1,2) or choose 3 to exit the game: ")
    return int(guess_input)
```

```python
def check_user_guess(list_values,user_input):
    if list_values[user_input]=='0':
        print("Good Job! You guessed it right!")
    else:
        print("Not right! try again!")
```

```python
list_val = ['0','','']
guess_input = -1

while guess_input != 3:
    guess_input = get_user_guess()
    if guess_input == 3:
        print("Exited the game!")
        break
    else:
        check_user_guess(shuffle_list(list_val),guess_input)
```
